importar bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

leitura dos dados armazenando csv em variavel

```python
df = pd. read_csv('lista_de_espera_sisu_2022_2.csv', sep='|', encoding='latin-1')
```

verificar leituras

```python
print(df.head())
print(df.columns)
```

exibir campos e status

```python
campus_por_estado = df ['UF_CAMPUS'].value_counts()
print(campus_por_estado)

cursos = df ['NOME_CURSO' ].value_counts ()
print(cursos)
```

criar grafico

```python
campus_por_estado.plot(kind='bar')
plt.xlabel('UF_CAMPUS' )
plt.title('Distribuição de Campus por Estado')
plt.show()
```

```python
cursos.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuição de Cursos')
plt.show()
```

```python


cursos2 = df ['NOME_CURSO' ].value_counts().nlargest(10)
print(cursos2)

```

```python
cursos2.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuição de Cursos')
plt.show()
```
