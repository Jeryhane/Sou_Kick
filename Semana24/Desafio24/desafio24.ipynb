importar bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

ler dados do csv

```python
df = pd. read_csv('chamada_regular_sisu_2022_2.csv', sep='|', encoding='latin-1')
```

testar dados

```python
print(df.head())
print(df.columns)
```

mostrar campos

```python
campus_por_estado = df ['UF_CAMPUS'].value_counts()
print(campus_por_estado)
```

```python
cursos = df ['NOME_CURSO' ].value_counts ()
print(cursos)
```

criar grafico

```python
campus_por_estado.plot(kind='bar')
plt.xlabel('UF_CAMPUS' )
plt.title('Distribuição de Campus por Estado')
plt.show()
```
